```python
import json

# Path to your text file containing the JSON-like data
input_file = "record.txt"
output_file = "formatted_data.json"

# Read the text file line by line
with open(input_file, "r") as f:
    lines = f.readlines()
print(lines)
# Parse each line as a JSON object and collect them into a list
data = [json.loads(line.strip()) for line in lines]

# Write the list of JSON objects to a proper JSON file
with open(output_file, "w") as f:
    json.dump(data, f, indent=4)

print(f"Formatted JSON saved to {output_file}")

```

```python
import pandas as pd

# Read the JSON stored in a text file
file_path = "formatted_data.json"

# Load the text file and parse as JSON
df = pd.read_json(file_path)

print(df.head())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(df["epoch"], df["loss"], marker="o", label="Loss Trend")

# Add labels, title, and legend
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Trend(Qwen) vs. Epoch")
plt.grid(True)
plt.legend()
plt.show()
```
